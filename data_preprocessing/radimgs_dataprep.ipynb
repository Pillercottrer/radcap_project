```python
import csv
import os
import json
import re
from pathlib import Path

out = []
out_bodypart_split = {}
out_bodypart_split['ankle'] = []
out_bodypart_split['elbow'] = []
out_bodypart_split['foot'] = []
out_bodypart_split['hip'] = []
out_bodypart_split['knee'] = []
out_bodypart_split['shoulder'] = []
out_bodypart_split['spine'] = []
out_bodypart_split['wrist'] = []
with open('/media/ehb/ssd_1tb/RadCapData/hasse_o_tage_classification_v4.csv', mode='r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    #csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    loc = '/media/ehb/ssd_1tb/RadCapData/PNG201709_small'
    for row in csv_reader:
        if line_count == 0:
            print(f'Column names are {", ".join(row)}')
            line_count += 1
        else:
            #print(f'\t{row[0]} works in the {row[1]} department, and was born in {row[2]}.')
            #print(row['Filename'])
            #print(row['Rprt'])
            
                
            
            split = row['Filename'].split(',')
            img_paths = []
            for img_path in split:
                img_path = img_path.replace('[', '')
                img_path = img_path.replace(']', '')
                img_path = img_path.replace("\'", '')
                img_path = loc + img_path
                
                my_file = Path(img_path)
                if my_file.is_file():
                    #print(img_path)
                    img_paths.append(img_path)
            if len(img_paths) == 0:
                continue
                    
                
            jimg = {}
            for key in row:
                jimg[key] = row[key]
            jimg['file_paths'] = img_paths
            if line_count < 10:
                print('Fracture? %s' % row['Fracture'])
                print('Fracture discplaced? %s' % row['Fracture_displaced'])
                print('Fracture undisplaced? %s' % row['Fracture_undisplaced'])
                print('Fracture intraarticular? %s' % row['Fracture_intraarticular'])
                print('Fracture spiral? %s' %row['Fracture_spiral'])
            """
            jimg['fracture'] = row['Fracture']
            jimg['fracture_displaced'] = row['Fracture_displaced']
            jimg['fracture_undisplaced'] = row['Fracture_displaced']
            jimg['fracture_intraarticular'] = row['Fracture_displaced']
            jimg['fracture_spiral'] = row['Fracture_displaced']
            jimg['fracture_comminute'] = row['Fracture_comminute']
            """
            
            
            sents = []
            sents.extend(row['Rprt'].split('.'))
            if line_count < 10:
                print(sents)
            if sents[-1] == '':
                sents = sents[:-1]
            if line_count < 10:
                print(sents)
                
            
            jimg['captions'] = sents
            
            if any("Ankle" in s for s in img_paths):
                out_bodypart_split['ankle'].append(jimg)
            elif any("Elbow" in s for s in img_paths):
                out_bodypart_split['elbow'].append(jimg)
            elif any("Foot" in s for s in img_paths):
                out_bodypart_split['foot'].append(jimg)
            elif any("Hip" in s for s in img_paths):
                out_bodypart_split['hip'].append(jimg)
            elif any("Knee" in s for s in img_paths):
                out_bodypart_split['knee'].append(jimg)
            elif any("Shoulder" in s for s in img_paths):
                out_bodypart_split['shoulder'].append(jimg)
            elif any("Spine" in s for s in img_paths):
                out_bodypart_split['spine'].append(jimg)
            elif any("Wrist" in s for s in img_paths):
                out_bodypart_split['wrist'].append(jimg)
                
                
            out.append(jimg)
            line_count += 1
    print(f'Processed {line_count} lines.')
    
    
print(out[0])
print(len(out))
print('Finished')
json.dump(out, open('radcap_data.json', 'w'))
#print(out_bodypart_split['ankle'])
print(len(out_bodypart_split['ankle']))
json.dump(out_bodypart_split, open('radcap_bodypartsplit_data.json', 'w'))
    
```
